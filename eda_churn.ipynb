### Transformação de Dados

> Coluna "TotalCharges" deveria ser do tipo `float`

```python
# Transformação da coluna "TotalCharges" para float (abordagem com astype)
try:
    df_contracts['TotalCharges'] = df_contracts['TotalCharges'].astype(float)
except Exception as e:
    print(f'Erro identificado: {e}.')
```

##### Disclaimer

```python
# Exemplo de quando o astype funciona
df = pd.DataFrame(
    {
        'id_venda': [1, 2, 3],
        'valor_venda': ['100.58', '199.99', '335.42']
    }
)
```

```python
df.info()
```

```python
# Conversão da coluna "valor_venda" para float
df['valor_venda'] = df['valor_venda'].astype(float)
```

```python
df.info()
```

```python
df
```

##### 2ª alternativa de transformação da coluna para `float`

```python
# Transformação da coluna "TotalCharges" para float (abordagem com to_numeric)
try:
    df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'])
except Exception as e:
    print(f'Erro identificado: {e}.')
```

```python
# Visualização do valor da linha 488 e da coluna "TotalCharges"
df_contracts.loc[488, 'TotalCharges']
```

##### 3ª alternativa

```python
# Transformação da coluna "TotalCharges" para float (abordagem com to_numeric com parâmetro errors='coerce')
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'], errors='coerce')
```

```python
# Visualização da linha 488 do dataframe de contratos
df_contracts.loc[488]
```

```python
# Filtro para identificar registros com valores nulos na coluna "TotalCharges"
df_contracts[df_contracts['TotalCharges'].isnull()]
```

> Após a transformação do tipo da coluna "TotalCharges", foi identificado 11 valores nulos

### Unificar DataFrames

```python
# Contagem da quantidade de linhas do dataframe de customers
len(df_customers)
```

```python
# Contagem da quantidade de linhas do dataframe de services
len(df_services)
```

```python
# Contagem da quantidade de linhas do dataframe de contracts
len(df_contracts)
```

```python
# Renomeação da coluna "customerID" para "IDCliente" no dataframe de services
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
# Visualização da estrutura do dataframe de services após a renomeação
df_services.info()
```

```python
# Unificação dos dataframes de customers e services
df_customers_services = df_customers.merge(df_services, on='IDCliente')
```

```python
# Visualização da estrutura do dataframe unificado
df_customers_services.info()
```

```python
# Visualização dos 3 primeiros registros do dataframe unificado
df_customers_services.head(3)
```

```python
# Unificação dos dataframes de customers_services e contracts (com nomes de colunas diferentes)
df_churn_temp = df_customers_services.merge(df_contracts, left_on='IDCliente', right_on='customerID')
```

```python
# Visualização da estrutura do dataframe resultante dos 3 primeiros
df_churn_temp.info()
```

```python
# Unificação dos 3 dataframes (customers, services e contracts) em um único dataframe
df_churn = df_customers.merge(df_services, on='IDCliente') \
           .merge(df_contracts, left_on='IDCliente', right_on='customerID')
```

```python
# Visualização da estrutura do dataframe resultante
df_churn.info()
```

```python
# Remoção da coluna "customerID" do dataframe de churn
df_churn.drop('customerID', axis=1, inplace=True) # axis=1 indica que a remoção é de uma coluna e axis=0 indicaria linha
```

```python
df_churn.info()
```

## Imputação de valores ausentes

```python
# Preenchimento de valores ausentes com ZERO
df_churn.fillna(0).query('TotalCharges == 0')
```

```python
# Preencimento de valores padrões a depender do tipo de dado
df_churn.fillna(value={
    'TotalCharges': 0,
    'Genero': 'Não Declarado'
}).query('TotalCharges == 0')
```

```python
# Preenchimento de valores ausentes com a média da coluna
media_TotalCharges = df_churn['TotalCharges'].mean()

df_churn.fillna(value={'TotalCharges': media_TotalCharges}).query('TotalCharges == @media_TotalCharges')
```

## Análise univariada

Hipótese 1: A faixa etária do cliente tem uma forte associação com o churn

```python
# Verificação de valores únicos em uma coluna
df_churn['Churn'].unique()
```

```python
# Contaagem de clientes usando a variável churn
df_churn['Churn'].value_counts()
```

```python
# Porcentagem de clientes que cancelaram o serviço e que não cancelaram
df_churn['Churn'].value_counts(normalize=True) * 100
```

```python
# Plotagem da distribuição de clientes que cancelaram o serviço e que não cancelaram
ax = df_churn['Churn'].value_counts().plot(kind='bar', title='Distribuição de clientes churn', xlabel='Churn', ylabel='Quantidade', color=['lightgreen', 'orange'], figsize=(10, 7))

# Adição de rótulos nos gráficos
ax.bar_label(ax.containers[0], label_type='edge', padding=3)
```

```python
# Plotagem da distribuição de clientes que cancelaram o serviço e que não cancelaram
ax = df_churn['Churn'].value_counts(normalize=True).plot(kind='bar', title='Distribuição de clientes churn', xlabel='Churn', ylabel='Quantidade', color=['lightgreen', 'orange'], figsize=(10, 7))

# Adição de rótulos nos gráficos
ax.bar_label(ax.containers[0], label_type='edge', padding=3)
```

```python
# Identificação dos tipos de contratos existentes
df_churn['Contract'].unique()
```

```python
# Plotagerm da distribuição de tipos de contratos
ax = df_churn['Contract'].value_counts().plot(kind='bar', title='Distribuição de tipos de contratos', xlabel='Tipo de contrato', ylabel='Quantidade', color=['lightblue', 'lightgreen', 'orange'], figsize=(10, 7))

# Adição de rótulos nos gráficos
ax.bar_label(ax.containers[0], label_type='edge', padding=3)
```

```python
# Plotagerm da distribuição de tipos de contratos
ax = df_churn['Contract'].value_counts(normalize=True).plot(kind='bar', title='Distribuição de tipos de contratos', xlabel='Tipo de contrato', ylabel='Quantidade', color=['lightblue', 'lightgreen', 'orange'], figsize=(10, 7))

# Adição de rótulos nos gráficos
ax.bar_label(ax.containers[0], label_type='edge', padding=3)
```

```python
# Plotagerm da distribuição de tipos de contratos para clientes que cancelaram
ax = df_churn.query('Churn == "Yes"')['Contract'].value_counts(normalize=True).plot(kind='bar', title='Distribuição de tipos de contratos', xlabel='Tipo de contrato', ylabel='Quantidade', color=['lightblue', 'lightgreen', 'orange'], figsize=(10, 7))

# Adição de rótulos nos gráficos
ax.bar_label(ax.containers[0], label_type='edge', padding=3)
```
