##### Avaliar a correlação entre 2 variáveis categóricas (qualitativas)

```python
# Executar um teste de hipótese chamado Chi-Square ou Qui-Quadrado de Pearson. Num teste de hipótese, 2 hipóteses são formuladas:
# A hipótese nula (H0): as variáveis são independentes
# E a hipótese complementar (H1): as variáveis são dependentes

# O teste serve para confirmar ou rejeitar a hipótese nula
# Se o p-valor for menor que 0.05, rejeitamos a hipótese nula e concluímos que as variáveis são dependentes

# Gerar um DataFrame com a tabela de contingência
df_crosstab_churn_contract = pd.crosstab(df_churn['Churn'], df_churn['Contract'])
```

```python
# Importar a função chi2_contingency do módulo scipy.stats
from scipy.stats import chi2_contingency
```

```python
# Calcular os scores do teste de hipótese e p-valor
chi_score_churn_contract = chi2_contingency(df_crosstab_churn_contract)
chi_score_churn_contract
```

```python
# Criando 2 séries com os valores de score e p-valor
chi_score = pd.Series(chi_score_churn_contract[0], name='Score')
p_value = pd.Series(chi_score_churn_contract[1], name='P-Value')
```

```python
# Apresentação de números com decimais sem notação científica
pd.set_option('display.float_format', lambda x: '%.30f' % x)
```

```python
# Criação do DataFrame com os resultados do teste de hipótese
df_chi_score_churn_contract = pd.DataFrame([chi_score, p_value], index=['Score', 'P-Value']).T
df_chi_score_churn_contract
```

> Diante do valor de p-valor (< 0.05), rejeitamos a hipótese nula e as variáveis `Chun` e `Contract` não são independentes. E pelo valor alto de Score (Qui-quadrado), elas possuem forte correlação

##### Avaliar a correlação entre uma variável qualitativa e quantitativa

Usar-se-á o mesmo teste de Chi-Square, mas transformar-se-á a variável quantitativa em qualitativa

```python
# Criação de uma coluna com base numa condição (TempoMenor6Meses)
import numpy as np

df_churn['TempoMenor6Meses'] = np.where(df_churn['tenure'] < 6, 'Yes', 'No')
```

```python
# Checar o dataframe com a nova coluna
df_churn.head()
```

```python
# Gerar a tabela de contingência entre as variáveis "Churn" e "TempoMenor6Meses"
df_crosstab_churn_tenure = pd.crosstab(df_churn['Churn'], df_churn['TempoMenor6Meses'])
```

```python
# Calcular os scores do teste de hipótese e p-valor
chi_score_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
# Criação de 2 séries com os valores de score e p-valor
chi_score_tenure = pd.Series(chi_score_churn_tenure[0], name='Score')
p_value_tenure = pd.Series(chi_score_churn_tenure[1], name='P-Value')
```

```python
# Criação do dataframe com os resultados do teste de hipótese
df_chi_score_churn_tenure = pd.DataFrame([chi_score_tenure, p_value_tenure], index=['Score', 'P-Value']).T
df_chi_score_churn_tenure
```

> Como o p-valor é menor que 0.05, rejeitamos a hipótese nula e concluímos que as variáveis são dependentes. Porém, o score é baixo, indicando que a relação entre Churn e TempoMenor6Meses é mais fraca que a relação entre Churn e Contract

### Z-Score (usado para distribuição normal)

Indica quantos desvios padrões um ponto específico de dados está distante da média

```python
# Z-score para identificar outliers (z-score = (x - mean) / std)
from scipy.stats import zscore

# Cálculo do z-score da coluna "TotalCharges" e tornando os valores absolutos
z_scores_TotalCharges_monthly = np.abs(zscore(df_churn_monthly['TotalCharges']))
z_scores_TotalCharges_monthly
```

```python
# Filtro para identificar outliers com z-score
df_churn_monthly[z_scores_TotalCharges_monthly > 3]
```
